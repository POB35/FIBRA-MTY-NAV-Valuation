# **3. Conceptos financieros clave utilizados en la valuación**

## **3.1 NOI (Net Operating Income) / ION (Ingreso Operativo Neto)**

El NOI o ION representa el ingreso generado por las propiedades después de gastos operativos, pero antes de intereses, impuestos, depreciación y amortización. Es una medida clave para evaluar la capacidad de generación de flujo de un portafolio inmobiliario y se utiliza comúnmente para valuar activos mediante tasas de capitalización.

## **3.2 Cap Rate (Tasa de capitalización)**

El cap rate es la tasa utilizada para convertir el NOI en el valor del activo inmobiliario. Se calcula como:

$$
Cap\ Rate= \frac {NOI}{Valor\ del\ activo}
$$

También puede interpretarse como el rendimiento esperado de una propiedad inmobiliaria. Un cap rate bajo implica una valuación más alta del activo, mientras que un cap rate alto implica una valuación más baja.

## **3.3 Cap Rate implícito de mercado**

Es la tasa de capitalización que el mercado está utilizando implícitamente para valuar el portafolio inmobiliario. Se calcula como:

$$
Cap\ Rate= \frac {NOI} {EV}
$$

donde:

$$
EV= Market\ Cap + Deuda\ Neta
$$

## **3.4 Valor Empresa (Enterprise Value, EV)**

El valor empresa representa el valor total del negocio, incluyendo tanto el capital de los accionistas como la deuda. En el contexto de una FIBRA, el EV puede aproximarse mediante:

$$
EV= \frac {NOI}{Cap\ Rate}
$$

## **3.5 Capitalización de mercado**

La capitalización de mercado representa el valor de mercado del capital de la empresa o FIBRA y se calcula como:

$$
Market\ Cap= Precio\ del\ CBFI \times CBFI's\ en\ circulación
$$

## **3.6 CBFI**

Los CBFIs (Certificados Bursátiles Fiduciarios Inmobiliarios) son los títulos que representan la participación de los inversionistas en el patrimonio de una FIBRA.

## **3.7 NAV (Net Asset Value)**

El NAV representa el valor neto de los activos atribuible a los tenedores de los certificados. Se calcula como:

$$
NAV= Valor\ de\ los\ activos - Deuda\ neta
$$

## **3.8 Deuda neta**

La deuda neta se calcula como la deuda total menos el efectivo disponible:

$$
Deuda\ neta= Deuda\ total - Efectivo
$$

Representa la obligación financiera real del fideicomiso.

## **3.9 Descuento/Prima vs NAV**

Mide la diferencia entre el valor intrínseco estimado (NAV por CBFI) y el precio de mercado:

$$
Descuento/Prima= \frac{NAV}{Precio} - 1
$$

El presente análisis tiene como objetivo estimar un rango de valor intrínseco para los certificados bursátiles fiduciarios inmobiliarios (CBFI) de Fibra Monterrey mediante un enfoque de capitalización directa del Ingreso Operativo Neto (NOI).

El modelo se basa en la anualización del ION más reciente reportado y en la aplicación de distintos escenarios de tasa de capitalización para estimar el valor del portafolio inmobiliario, el NAV ajustado y el NAV por CBFI.

- El NOI anual se aproxima mediante la anualización del ION del trimestre más reciente.
- El portafolio inmobiliario se valúa mediante un modelo de capitalización directa (NOI / Cap Rate).
- La deuda neta se mantiene constante.
- No se consideran adquisiciones, desinversiones o desarrollos futuros.
- No se modela crecimiento de rentas ni inflación de manera explícita.
- El análisis se enfoca en el valor del portafolio inmobiliario y no en otras posibles fuentes de valor.

Por lo tanto, los resultados obtenidos deben interpretarse como una estimación aproximada del valor intrínseco bajo distintos escenarios de tasa de capitalización, y no como una valuación exacta.

# **5. Metodología**

La valuación de Fibra Monterrey se realiza mediante un enfoque de NAV ajustado basado en la capitalización del Ingreso Operativo Neto (ION/NOI). Este enfoque es comúnmente utilizado en la valuación de vehículos inmobiliarios como FIBRAs, ya que permite estimar el valor económico del portafolio inmobiliario a partir de su capacidad de generación de ingresos.

La metodología del proyecto se desarrolla en los siguientes pasos:

1. Tomar el ION trimestral más reciente reportado por Fibra Monterrey.
2. Anualizar dicho ION para aproximar un NOI run-rate.
3. Definir escenarios de cap rate.
4. Calcular el Valor Empresa (EV) mediante la fórmula:

$$
EV = \frac{NOI}{Cap\ Rate}
$$

5. Obtener el NAV ajustado restando la deuda neta:

$$
NAV = EV - Deuda\ neta
$$


6. Calcular el NAV por CBFI:

$$
NAV\ por\ CBFI = \frac{NAV}{CBFIs\ en\ circulación}
$$

7. Comparar el NAV por CBFI con el precio de mercado actual para determinar si el certificado cotiza con descuento, valor justo o prima.
8. Calcular el cap rate implícito de mercado para validar la valuación observada y comparar la tasa de capitalización que descuenta el mercado contra los escenarios utilizados en el modelo.

Este procedimiento permite estimar un rango de valor intrínseco aproximado y evaluar la valuación actual del certificado.

Para desarrollar el modelo se utilizan principalmente las librerías `pandas` para manipulación de datos y `matplotlib` para visualización de resultados.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## **7.2 Definición de variables base**

A continuación se definen las variables principales del modelo con base en la información reportada por Fibra Monterrey al cierre del 4T-2025.

```python
# Datos base reportados
ion_4t25_miles = 768_562                    # ION trimestral en miles de pesos
efectivo_millones = 3_377.1                 # millones de pesos
deuda_total_millones = 11_130.7             # millones de pesos
cbfis_en_circulacion_miles = 2_438_674.664  # en miles
precio_mercado = 15.24                      # precio por CBFI en pesos
reserva_territorial = 615_000_000           # millones de pesos
otros_activos_exc_efectivo_e_inversiones_financieras = 561_000_000 # millones de pesos

# Conversión de unidades
ion_4t25 = ion_4t25_miles * 1_000
efectivo = efectivo_millones * 1_000_000
deuda_total = deuda_total_millones * 1_000_000
cbfis = cbfis_en_circulacion_miles * 1_000
```

## **7.3 Cálculo del ION anualizado y deuda neta**

El ION anualizado se obtiene multiplicando el ION trimestral más reciente por 4 (trimestres). Posteriormente, se calcula la deuda neta como la diferencia entre la deuda total y el efectivo disponible.

## **7.4 Definición de escenarios de cap rate**

Para efectos de la valuación, se consideran dos tipos de tasas de capitalización:

1. **Referencias de mercado: tasas implícitas calculadas a partir del valor de mercado actual de la FIBRA.**
   
2. **Escenarios de valuación: tasas de capitalización utilizadas como supuestos para estimar el valor intrínseco bajo distintos niveles de rendimiento requerido por los inversionistas.**

Las tasas consideradas son las siguientes:

```python
cap_rates = {
    "Referencia de mercado": {
        "Cap rate implícito a EV total": 0.0684,
        "Cap rate implícito a propiedades operativas": 0.0703
    },
    "Escenario de valuación": {
        "Escenario Base": 0.0750,
        "Escenario Conservador": 0.0800
    }
}

filas = []
for tipo, escenarios in cap_rates.items():
    for nombre, tasa in escenarios.items():
        filas.append([tipo, nombre, tasa])

df_cap_rates = pd.DataFrame(filas, columns=["Tipo", "Escenario", "Cap Rate"])
df_cap_rates["Cap Rate"] = df_cap_rates["Cap Rate"].map(lambda x: f"{x:.2%}")

df_cap_rates.style.hide(axis="index").set_table_styles([
    {'selector': 'th', 'props': [('text-align', 'center'), ('font-weight', 'bold')]},
    {'selector': 'td', 'props': [('text-align', 'center')]},
    {'selector': 'table', 'props': [('margin', '0 auto')]},
    {'selector': 'td, th', 'props': [('padding', '6px 12px')]}
])
```

Las tasas implícitas de mercado **(6.84% y 7.03%)** representan el rendimiento al cual el mercado está valuando actualmente el portafolio de Fibra Monterrey.

Por otro lado, los escenarios base y conservador **(7.50% y 8.00%)** representan tasas de capitalización más altas, que reflejan supuestos de valuación más conservadores y mayores rendimientos requeridos por los inversionistas.

El análisis de sensibilidad del valor del certificado se realizará utilizando estas tasas de capitalización, con el objetivo de estimar un rango de valor intrínseco y compararlo contra el precio de mercado actual.

## **7.5 Cálculo del Valor Empresa (EV), NAV y NAV por CBFI**

Con base en cada escenario de cap rate, se estima el valor del portafolio (EV), posteriormente el NAV ajustado y finalmente el NAV por CBFI.

El procedimiento es el siguiente:

1. Se anualiza el ION más reciente para estimar un NOI run-rate.
2. Se calcula el Valor Empresa (EV) dividiendo el NOI entre el cap rate.
3. Se calcula el NAV ajustado restando la deuda neta al Valor Empresa.
4. Se calcula el NAV por CBFI dividiendo el NAV entre los CBFIs en circulación.
5. Finalmente, se compara el NAV por CBFI contra el precio de mercado actual para estimar la prima o descuento.

Las fórmulas utilizadas son:

$$
EV = \frac{NOI}{Cap\ Rate}
$$


$$
NAV = EV - Deuda\ Neta
$$


$$
NAV/CBFI = \frac{NAV}{CBFIs\ en\ circulación}
$$


$$
Prima/Descuento = \frac{NAV/CBFI}{Precio\ de\ mercado} - 1
$$


```python
# Escenarios de cap rate
escenarios = {
    "Mercado EV total": 0.0684,
    "Mercado Propiedades": 0.0703,
    "Escenario Base": 0.0750,
    "Escenario Conservador": 0.0800
}

resultados = []

for nombre, cap_rate in escenarios.items():
    
    EV = noi_anual / cap_rate
    NAV = EV - deuda_neta
    NAV_por_CBFI = NAV / cbfis
    prima_descuento = (NAV_por_CBFI / precio_mercado) - 1
    
    resultados.append([
        nombre,
        cap_rate,
        EV,
        NAV,
        NAV_por_CBFI,
        precio_mercado,
        prima_descuento
    ])

df_resultados = pd.DataFrame(resultados, columns=[
    "Escenario",
    "Cap Rate",
    "EV",
    "NAV",
    "NAV/CBFI",
    "Precio",
    "Prima/Descuento"
])

```

## **7.7 Cálculo del cap rate implícito**

# **8. Discusión**

## **8.1  Cap rate implícito vs cap rate de valuación**

El cap rate implícito a valor empresa se estima en aproximadamente 6.84%, mientras que el cap rate implícito ajustado para activos operativos se ubica alrededor de 7.03%. Ambos valores se encuentran por debajo de los cap rates utilizados en los escenarios base y conservador (7.75%–8.25%), lo cual indica que el mercado está valuando el portafolio a una tasa de capitalización baja y, por lo tanto, a una valuación relativamente alta.

En el sector inmobiliario, una menor tasa de capitalización implica un mayor valor del activo, por lo que el precio actual del certificado parece reflejar expectativas favorables sobre la calidad del portafolio, la estabilidad de los flujos de efectivo y las perspectivas del segmento industrial en México. Sin embargo, esta valuación también implica que el certificado podría ser sensible a incrementos en las tasas de capitalización del mercado, lo que podría impactar negativamente el valor estimado del NAV en el futuro.

## **8.2  Factores que pueden justificar un cap rate bajo**

Existen diversos factores que podrían justificar que el mercado asigne a Fibra Monterrey una tasa de capitalización menor a la utilizada en los escenarios base y conservador. Entre estos factores destacan:

- Alta tasa de ocupación del portafolio.
- Contratos de arrendamiento de largo plazo.
- Ingresos mayoritariamente denominados en dólares.
- Exposición predominante al segmento industrial, el cual se ha beneficiado del fenómeno de nearshoring en México.
- Percepción positiva del mercado respecto al gobierno corporativo y la administración de la FIBRA.


## **8.3  Riesgos del modelo y sensibilidad al cap rate**

El modelo de valuación utilizado es altamente sensible a la tasa de capitalización. Pequeños cambios en el cap rate generan variaciones importantes en el valor estimado del NAV por CBFI.

Por lo tanto, uno de los principales riesgos para la valuación de Fibra Monterrey es un posible incremento en las tasas de capitalización del mercado, lo cual podría derivarse de incrementos en tasas de interés, cambios en la percepción de riesgo del sector inmobiliario o cambios en las condiciones económicas. En un escenario donde el cap rate de mercado aumente, el valor estimado del portafolio y el NAV por CBFI tenderían a disminuir.

## **8.4  Interpretación general**

En conjunto, los resultados del modelo sugieren que el precio actual de Fibra Monterrey es consistente con un escenario de cap rate bajo, lo que implica que el mercado ya reconoce la calidad del portafolio y las perspectivas del segmento industrial. No obstante, bajo supuestos de cap rate más conservadores, el valor estimado por CBFI resulta inferior al precio de mercado, lo que indica que el potencial de apreciación podría ser limitado si las condiciones del mercado cambian y las tasas de capitalización aumentan.

# **9. Limitaciones del modelo**

El presente análisis tiene diversas limitaciones que deben considerarse al interpretar los resultados obtenidos:

1. **Uso de NOI anualizado (run-rate)**  
   El modelo utiliza el NOI del cuarto trimestre anualizado como aproximación del NOI anual. Este supuesto implica que el desempeño operativo futuro será similar al observado en el trimestre más reciente, lo cual podría no cumplirse si cambian las condiciones de ocupación, las rentas o se incorporan nuevos inmuebles al portafolio.

2. **Sensibilidad del modelo al cap rate**  
   El valor estimado del portafolio y el NAV por CBFI dependen de manera significativa de la tasa de capitalización utilizada. Pequeñas variaciones en el cap rate generan cambios importantes en el valor estimado, por lo que los resultados deben interpretarse como un rango de valores bajo distintos escenarios y no como un valor puntual.

3. **No se modela crecimiento del NOI**  
   El modelo asume implícitamente que el NOI se mantiene constante en el tiempo, por lo que no se consideran incrementos futuros en rentas, renovaciones de contratos, inflación, expansiones o nuevas adquisiciones que podrían incrementar el NOI en el futuro.

4. **Uso de deuda neta como aproximación del pasivo financiero**  
   Para el cálculo del NAV se utilizó la deuda neta reportada, lo cual es una aproximación razonable; sin embargo, cambios en el tipo de cambio, refinanciamientos o nuevas emisiones de deuda podrían modificar el nivel de apalancamiento y, por lo tanto, el NAV.

5. **No se consideran otros activos o pasivos a valor de mercado**  
   El modelo se enfoca principalmente en el valor del portafolio inmobiliario, por lo que otros activos o pasivos del balance se consideran a valor en libros y no necesariamente a valor de mercado.

6. **El modelo no incorpora el costo de capital ni un análisis de flujos descontados**  
   La valuación se realiza mediante un enfoque de capitalización directa (NOI / Cap Rate) y no mediante un modelo de flujos descontados (DCF), el cual permitiría modelar de manera más detallada el crecimiento esperado del NOI, el costo de la deuda y el costo del capital.

Por estas razones, los resultados del modelo deben interpretarse como una estimación aproximada del valor intrínseco bajo distintos escenarios de tasa de capitalización.

# **10. Conclusiones**

El objetivo del presente análisis fue estimar un rango de valor intrínseco aproximado para los CBFIs de Fibra Monterrey mediante un modelo de NAV ajustado basado en la capitalización del NOI y distintos escenarios de tasa de capitalización, con el fin de evaluar si el precio de mercado refleja un descuento o una prima respecto a dicho valor.

Los resultados muestran que el cap rate implícito de mercado se ubica aproximadamente entre 6.8% y 7.0%, lo cual representa una tasa de capitalización baja en comparación con los cap rates utilizados en los escenarios base y conservador (7.5% y 8.0%). Esto implica que el mercado actualmente valúa el portafolio inmobiliario a una tasa de rendimiento baja y, por lo tanto, a una valuación relativamente alta.

Al comparar el NAV estimado por CBFI bajo los distintos escenarios contra el precio de mercado actual, se observa que el certificado cotiza cercano a su valor razonable únicamente bajo los cap rates implícitos de mercado. Sin embargo, bajo escenarios de valuación más conservadores, el NAV estimado por CBFI resulta inferior al precio de mercado, lo que sugiere que el certificado podría cotizar con prima bajo dichos supuestos.

**Por lo tanto, en respuesta a la pregunta de investigación, el análisis sugiere que el precio actual de mercado del CBFI de Fibra Monterrey no refleja un descuento claro frente a su valor intrínseco estimado mediante NAV ajustado. Más bien, el mercado parece reconocer plenamente el valor del portafolio e incluso valuarlo bajo supuestos optimistas de tasa de capitalización.**

**En conclusión, el precio actual del certificado parece reflejar un escenario favorable en términos de tasa de capitalización, por lo que el potencial de apreciación del CBFI podría depender en mayor medida del crecimiento futuro del NOI, nuevas adquisiciones, expansiones del portafolio o mejoras en las condiciones del mercado, más que de una expansión adicional en múltiplos de valuación.**
